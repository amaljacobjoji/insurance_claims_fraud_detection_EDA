#### Visualising using regplot comparing Fraud detected based on the policy holders age 

```python
policyAge = df.groupby('age')['fraud_reported'].sum()
plt.figure(figsize=(20,5))
plt.title("Policy Holder's Age Vs Frauds detected")

ax = sns.regplot(x=policyAge.index,y=policyAge.values,color='black')
```

#### DISPLOT OF COLLISION TYPE VS TOTAL AMOUNT CLAIMED ON VEHICLE

```python
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.displot([df['collision_type'],df['vehicle_claim']],palette= 'Paired_r')
plt.title('collision type vs Vehicle Claim', fontsize = 20)
plt.show()
```

```python
fraud_rate = df['fraud_reported'].mean() * 100
print(f"Fraudulent claims: {fraud_rate:.2f}%")
```

#### COMPARE  CLAIM DISTRIBUTIONS FOR FRAUDULENT AND NON-FRAUDULENT CLAIMS

```python
df[df['fraud_reported'] == True]['total_claim_amount'].hist(alpha=0.5, label='Fraudulent')
df[df['fraud_reported'] == False]['total_claim_amount'].hist(alpha=0.5, label='Non-fraudulent')
plt.legend()
```

##### VIOLINPLOT  SHOWING DATA BASED ON PERCENTAGE OF FRAUDULENT CLAIMS BY POLICY DURATION(DIFFERENCE BETWEEN POLICY STARTED DATE AND DATE OF INCIDENT HAPPENED)

```python
plt.figure(figsize=(12, 6))
adata=df.groupby("difference_period")["fraud_reported"].mean()
sns.violinplot(df,x=adata,color='brown',inner="stick",linecolor="k")
plt.title("Percentage of Fraudulent Claims by Policy Duration")
plt.xlabel("difference_period")
plt.ylabel("Proportion of Fraudulent Claims")
plt.show()
```

#### PIE CHART PRESENTATION OF AUTO MAKER OF VEHICLE AND TOTAL AMOUNT CLAIMED ON VEHCLE FOR THE INCIDENT HAPPENED

```python
df.groupby(['auto_make']).sum().plot( kind='pie', y='vehicle_claim',autopct='%1.1f%%')
plt.show()
```
